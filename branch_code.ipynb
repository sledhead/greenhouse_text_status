```python

def look_for_Key(dic,key):
  if key in dic.keys():
    print("Present, value =", dic[key])
  else:
    print("Not present")


test_dict = {'J-13':'snow', 'J-23':'light', 'J-2':'mikl', 'J-55':'sea', 'J-7':'japan'}

look_for_Key(test_dict,'J-23')
```

```python
look_for_Key(test_dict,'J-5')
```

```python
for i_num in range(1,60,1):

  #build possible key string
  find_key_str = f'J-{i_num}'

  #call function to determine if requested key str exists.
  look_for_Key(test_dict,find_key_str)
  
```

```python
vv = ['J-13', 'J-23', 'J-2', 'J-55', 'J-7']
print(type(vv))
print(vv.sort())
print(vv)

def special_sort(ele):
  #take the string passed in and return the number.
  #split the letter from the number and return the num
  temp_lst = ele.split('-')
  temp_num = int(temp_lst[1])
  return temp_num
```

```python
vv.sort(key=special_sort)
print(vv)
```
